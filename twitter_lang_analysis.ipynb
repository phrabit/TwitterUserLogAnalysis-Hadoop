```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

lang = pd.read_csv('C:/db_web/lang_result.csv')

# Concatenate all the languages into a single string
text = ' '.join(lang['lang'])

# Create the word cloud object
wordcloud = WordCloud(width=400, height=400, background_color='white').generate(text)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Read the CSV file
df = pd.read_csv('C:/db_web/lang_result.csv')

# Concatenate all the languages into a single string
text = ' '.join(df['lang'])

# Create the word cloud configuration
wordcloud = WordCloud(
    width=400,
    height=400,
    background_color='white',
    colormap='viridis',  # Change the colormap
    max_words=100,  # Set the maximum number of words in the word cloud
    max_font_size=100,  # Set the maximum font size
    random_state=42  # Set a random state for reproducibility
).generate(text)

# Plot the word cloud
plt.figure(figsize=(12, 6))  # Adjust the figure size
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.tight_layout()  # Add padding between the plot and the edge of the figure
plt.show()
```
