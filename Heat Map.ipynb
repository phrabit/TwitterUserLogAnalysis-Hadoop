```python
import pandas as pd
from geopy.geocoders import Nominatim

# 데이터프레임 생성
data = {
    'location': [
        'Orlando, Florida',
        'Community of Madrid, Spain',
        'Tucson, AZ',
        'Hajipur, India',
        'Guwahati, India',
        'Bengaluru, India',
        'Dhaka, Bangladesh',
        'Philadelphia, PA',
        'Granite Falls, NC',
        'Texas, USA',
        'Newport Beach, CA',
        'Urbandale, IA',
        'Houston, TX',
        'West Covina, CA',
        'Lagos, Nigeria',
        'Portland, Oregon'
    ]
}

df = pd.DataFrame(data)

# Nominatim 지오코딩 서비스 객체 생성
geolocator = Nominatim(user_agent="my-custom-agent")

# 'location' 열의 지역이름을 경도와 위도로 변환하여 새로운 열로 추가
df['coordinates'] = df['location'].apply(lambda x: geolocator.geocode(x).point if geolocator.geocode(x) is not None else None)

# 결과 출력
print(df)
```

```python
import folium
from folium.plugins import HeatMap

# Draw a basemap
maps = folium.Map(location=[37.58, 127.0], tiles='openstreetmap', zoom_start=11)

# Add heatmap
HeatMap(
    data=df[['coordinates']], 
    radius=20,
).add_to(maps)

# Display the map
maps
```

```python
from geopy.point import Point

# 호출 결과를 캐시하는 데코레이터 설정
def geocode_address(address):
    return geolocator.geocode(address)

# 'location' 열의 지역이름을 경도와 위도로 변환하여 새로운 열로 추가
df['coordinates'] = df['location'].apply(lambda x: geocode_address(x).point if geocode_address(x) is not None else None)
df['latitude'] = df['coordinates'].apply(lambda x: x.latitude if x is not None else None)
df['longitude'] = df['coordinates'].apply(lambda x: x.longitude if x is not None else None)

# 결과 출력
print(df)
```

```python
# Draw a basemap
maps = folium.Map(location=[37.58, 127.0], tiles='openstreetmap', zoom_start=11)

# Add heatmap
HeatMap(
    data=df[['latitude', 'longitude']], 
    radius=20,
).add_to(maps)

# Display the map
maps
```
